```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
ingestion_files = {'customers' : '/Volumes/ecom/bronze/raw_data/olist_customers_dataset.csv',
                  'geolocation' : '/Volumes/ecom/bronze/raw_data/olist_geolocation_dataset.csv',
                  'order_items':'/Volumes/ecom/bronze/raw_data/olist_order_items_dataset.csv',
                  'order_payments':'/Volumes/ecom/bronze/raw_data/olist_order_payments_dataset.csv',
                  'order_reviews':'/Volumes/ecom/bronze/raw_data/olist_order_reviews_dataset.csv',
                  'orders':'/Volumes/ecom/bronze/raw_data/olist_orders_dataset.csv',
                  'products':'/Volumes/ecom/bronze/raw_data/olist_products_dataset.csv',
                  'sellers':'/Volumes/ecom/bronze/raw_data/olist_sellers_dataset.csv',
                  'product_category_name_translation':'/Volumes/ecom/bronze/raw_data/product_category_name_translation.csv'}
```

```python
for key, value in ingestion_files.items():
    df = spark.read.format('csv')\
                    .option('header','True')\
                    .load(value)
    
    df = df.withColumn('_ingestion_time', current_timestamp()).withColumn('_source_file', lit(value))

    df.write.format('delta')\
            .mode('append')\
            .saveAsTable(f'ecom.bronze.{key}')
```
